```python
import numpy as np
from nptdms import TdmsFile

# Add the FILENAME at the end of the path
file_path = r"C:\Users\satres\Documents\tdms\TDMS_Analysis\first.tdms"
#                                                          ^^^^^^^^^^^
#                                                          file name added here

tdms = TdmsFile.read(file_path)
group = tdms["SIGNAUX"]

# Get gains from group properties
GAIN_U = 200.0
GAIN_I = 10.0

# Get raw data
U0_raw = group["U0"][:]
I0_raw = group["I0"][:]

# === VOLTAGE CHECK ===
print("=== VOLTAGE CHECK ===")
print(f"Raw U0 statistics (in Volts):")
print(f"  Min: {U0_raw.min():.4f} V")
print(f"  Max: {U0_raw.max():.4f} V")
print(f"  RMS: {np.sqrt(np.mean(U0_raw**2)):.4f} V")

U0_rms_raw = np.sqrt(np.mean(U0_raw**2))
print(f"\nInterpretation A (raw / gain): RMS = {U0_rms_raw / GAIN_U:.4f} V_actual")
print(f"Interpretation B (raw * gain): RMS = {U0_rms_raw * GAIN_U:.4f} V_actual")

print("\nFor a 230V phase / 400V line system, you should see RMS = 230 V phase.")
print("Whichever interpretation gives ~230 V is the correct one.")

# === CURRENT CHECK ===
print("\n=== CURRENT CHECK ===")
I0_rms = np.sqrt(np.mean(I0_raw**2))
print(f"Raw I0 RMS (in Volts): {I0_rms:.4f} V")
print(f"\nInterpretation A (raw / gain): RMS = {I0_rms / GAIN_I:.4f} A_actual")
print(f"Interpretation B (raw * gain): RMS = {I0_rms * GAIN_I:.4f} A_actual")

print("\nFor TA 56-24 (rated 0.34 A), you should see < 0.5 A at rated load.")
```

```python
import os
import numpy as np
import pandas as pd
from nptdms import TdmsFile

# === CONFIGURATION ===
folder_path = r"C:\Users\satres\Documents\tdms\TDMS_Analysis"
GAIN_U = 200.0
GAIN_I = 10.0

# === FIND ALL TDMS FILES ===
tdms_files = []
for f in os.listdir(folder_path):
    if f.endswith(".tdms") and not f.endswith("_index.tdms"):
        tdms_files.append(os.path.join(folder_path, f))

print(f"Found {len(tdms_files)} TDMS files:\n")
for f in tdms_files:
    print(f"  - {os.path.basename(f)}")

# === ANALYZE EACH FILE ===
results = []

for file_path in tdms_files:
    filename = os.path.basename(file_path)
    print(f"\n{'='*70}")
    print(f"ANALYZING: {filename}")
    print(f"{'='*70}")
    
    try:
        tdms = TdmsFile.read(file_path)
        group = tdms["SIGNAUX"]
        
        # Get all 6 channels
        U0 = group["U0"][:]
        U1 = group["U1"][:]
        U2 = group["U2"][:]
        I0 = group["I0"][:]
        I1 = group["I1"][:]
        I2 = group["I2"][:]
        
        # Sample rate from metadata
        dt = group["U0"].properties["wf_increment"]
        fs = 1.0 / dt
        duration = len(U0) * dt
        
        print(f"  Samples: {len(U0):,}")
        print(f"  Sample rate: {fs:.0f} Hz")
        print(f"  Duration: {duration:.1f} s ({duration/60:.1f} min)")
        
        # === VOLTAGE ANALYSIS ===
        # Try both interpretations
        U0_rms_raw = np.sqrt(np.mean(U0**2))
        U1_rms_raw = np.sqrt(np.mean(U1**2))
        U2_rms_raw = np.sqrt(np.mean(U2**2))
        
        print(f"\n  --- VOLTAGE (raw RMS in stored Volts) ---")
        print(f"    U0 RMS: {U0_rms_raw:.4f}    U1 RMS: {U1_rms_raw:.4f}    U2 RMS: {U2_rms_raw:.4f}")
        
        print(f"\n    Interpretation A (raw × gain={GAIN_U}):")
        print(f"      U0: {U0_rms_raw * GAIN_U:.1f} V    U1: {U1_rms_raw * GAIN_U:.1f} V    U2: {U2_rms_raw * GAIN_U:.1f} V")
        print(f"    Interpretation B (raw / gain={GAIN_U}):")
        print(f"      U0: {U0_rms_raw / GAIN_U:.4f} V    U1: {U1_rms_raw / GAIN_U:.4f} V    U2: {U2_rms_raw / GAIN_U:.4f} V")
        
        # === CURRENT ANALYSIS ===
        I0_rms_raw = np.sqrt(np.mean(I0**2))
        I1_rms_raw = np.sqrt(np.mean(I1**2))
        I2_rms_raw = np.sqrt(np.mean(I2**2))
        
        I0_peak_raw = np.max(np.abs(I0))
        
        print(f"\n  --- CURRENT (raw RMS in stored Volts) ---")
        print(f"    I0 RMS: {I0_rms_raw:.4f}    I1 RMS: {I1_rms_raw:.4f}    I2 RMS: {I2_rms_raw:.4f}")
        print(f"    I0 peak (abs): {I0_peak_raw:.4f}")
        
        print(f"\n    Interpretation A (raw × gain={GAIN_I}):")
        print(f"      I0: {I0_rms_raw * GAIN_I:.4f} A RMS    Peak: {I0_peak_raw * GAIN_I:.4f} A")
        print(f"    Interpretation B (raw / gain={GAIN_I}):")
        print(f"      I0: {I0_rms_raw / GAIN_I:.4f} A RMS    Peak: {I0_peak_raw / GAIN_I:.4f} A")
        
        # Save for summary table
        results.append({
            'file': filename,
            'duration_s': duration,
            'U0_rms_×gain': U0_rms_raw * GAIN_U,
            'U0_rms_÷gain': U0_rms_raw / GAIN_U,
            'I0_rms_×gain': I0_rms_raw * GAIN_I,
            'I0_rms_÷gain': I0_rms_raw / GAIN_I,
            'I0_peak_×gain': I0_peak_raw * GAIN_I,
            'I0_peak_÷gain': I0_peak_raw / GAIN_I,
        })
        
    except Exception as e:
        print(f"  ERROR processing {filename}: {e}")

# === SUMMARY TABLE ===
print(f"\n\n{'='*70}")
print("SUMMARY ACROSS ALL FILES")
print(f"{'='*70}\n")

if results:
    df = pd.DataFrame(results)
    print(df.to_string(index=False))
    
    print("\n\n=== INTERPRETATION GUIDE ===")
    print("Look at the U0 columns:")
    print("  - If U0_rms_×gain values are around 200-250 V → multiply by gain (Interpretation A)")
    print("  - If U0_rms_÷gain values are around 200-250 V → divide by gain (Interpretation B)")
    print("\nFor a 230 V (phase-to-neutral) or 400 V (line-to-line) European supply,")
    print("the correct interpretation gives values matching one of these standards.")
    print("\nThen apply the SAME interpretation (×gain or ÷gain) to the currents.")
    print("For a Techtop TA 56-24, expect:")
    print("  - Idle/light load: 0.1-0.4 A RMS")
    print("  - Rated load: ~0.34 A RMS")
    print("  - Inrush peak: 1.5-2.5 A peak")
```
